## 加载数据

```python
import pandas as pd

# df = pd.read_html("https://olympics.com/tokyo-2020/olympic-games/zh/results/all-sports/medal-standings.htm")[0]

path = "/content/drive/MyDrive/Pandas_Advanced_Exercise-main/5 - 数据筛选与修改/东京奥运会奖牌数据.csv"
df = pd.read_csv(path)  # 如果未联网可以使用这条命令加载数据
df.head()
```

## 5-1 数据修改



### 1- <font color = '#FB8E00'>数据修改</font>｜列名

<br>

将原 `df` 列名 `Unnamed: 2`、`Unnamed: 3`、`Unnamed: 4` 修改为 `金牌数`、`银牌数`、`铜牌数`

```python
df = df.rename(columns={"Unnamed: 2":"金牌数", "Unnamed: 3":"银牌数", "Unnamed: 4":"铜牌数"})
df.head()
```

### 2 - <font color = '#FB8E00'>数据修改</font>｜行索引

<br>

将第一列（排名）设置为索引

```python
df.set_index(['排名'], inplace=True)
df.head()
```

### 3 - <font color = '#FB8E00'>数据修改</font>｜修改索引名

<br>

修改索引名为 金牌排名

```python
# df.set_index(name='金牌排名')
# df.head()
```

```python
df.rename_axis("金牌排名",inplace=True)
df.head()
```

### 4 - <font color = '#FB8E00'>数据修改</font>｜修改值

<br>

将 ROC（第一列第五行）修改为 俄奥委会

```python
df.iloc[4, 0] = ' 俄奥委会'  #index不计数
df.head()
```

### 5 - <font color = '#FB8E00'>数据修改</font>｜替换值（单值）

<br>

将金牌数列的数字 `0` 替换为 `无`

```python
df['金牌数'].replace(0,'无',inplace=True)
df.head()
```

### 6 - <font color = '#FB8E00'>数据修改</font>｜替换值（多值）



同时替换

- 将 `无` 替换为 缺失值
- 将 0 替换为 `None`

注意：缺失值的 Nan 该怎么生成？

```python
import numpy as np
df.replace(['无',0],[np.nan,'None'],inplace = True)
```

### 7 - 数据查看

查看各列数据类型

```python
df.info()
```

### 8 - <font color = '#FB8E00'>数据修改</font>｜修改类型

将 `金牌数` 列类型修改为 `int`

```python
# df['金牌数'] = df['金牌数'].astype(int)
# df.info()
```

```python
df['金牌数'] = df['金牌数'].fillna(0).astype(int)
df.info()
```

### 9 - <font color = '#1B85FF' >数据增加</font>｜新增列（固定值）

<br>

**重新加载数据** 并 **新增一列** 比赛地点，值为 `东京`


```python
df['比赛地点'] = '东京'
df.head()
```

### 10 - <font color = '#1B85FF' >数据增加</font>｜新增列（计算值）

<br>

**新增一列** 金银牌总数列，值为该国家金银牌总数

```python
df['银牌数'].replace(['None', 0], inplace=True)
# df['银牌数'] = df['银牌数'].fillna(0).astype(int)
df['金银牌总数列'] = df['金牌数'] + df['银牌数']
df.head()
```

### 11 - <font color = '#1B85FF' >数据增加</font>｜新增列（比较值）

新增一列 最多奖牌数量 列，值为该过金银牌数量种最多的一个奖牌数量

例如美国银牌最多，则为41，中国为38

```python
# df['最多奖牌数量'] = max(df['金牌数'], df['银牌数'], df['铜牌数'])
# df.head()
```

```python
# df['最多奖牌数量'] = max(df[["金牌数", "银牌数",'铜牌数']])
# df.head()
```

```python
df['铜牌数'].replace(['None', 0], inplace=True)
df['最多奖牌数量'] = df.bfill(1)[["金牌数", "银牌数",'铜牌数']].max(1)
df.head()
```

### 12 - <font color = '#1B85FF' >数据增加</font>｜新增列（判断值）

新增一列 金牌大于30

如果一个国家的金牌数大于 30 则值为 是，反之为 否

```python
# df['金牌大于30'] = [1 if df["金牌数"] > 30 else 0]
# df.head()
```

```python
df['金牌大于30']  = np.where(df['金牌数'] > 30, '是', '否')
df.head()
```

### 13 - <font color = '#1B85FF' >数据增加</font>｜增加多列

新增两列，分别是

- 金铜牌总数（金牌数+铜牌数）
- 银铜牌总数（银牌数+铜牌数）

```python
df['金铜牌总数'] = df['金牌数'] + df['铜牌数']
df['银铜牌总数'] = df['银牌数'] + df['铜牌数']
df.head()
```

### 14 - <font color = '#1B85FF' >数据增加</font>｜新增列（引用变量）

新增一列金牌占比，为各国金牌数除以总金牌数（gold_sum）

```python
df['金牌占比'] = df['金牌数'] / df['金铜牌总数'].sum()
df.head()
```

<br>

在 df 末尾追加一行，内容为 0,1,2,3... 一直到 `df` 的列长度

```python
# l = [i for i in range(len(df.columns))]  ## 注意：使用列表创建df时是按照列来进行创建的，这里是需要每一行的数值
# print(l)
# df0 = pd.DataFrame(l, columns=df.columns)
# df0
# # df2 = df.append(df0)
# # df2
```

```python
# lll = np.array([i for i in range(len(df.columns))]).T  ## 注意：使用列表创建df时是按照列来进行创建的，这里是需要每一行的数值
# print(lll.shape)
# df0 = pd.DataFrame(lll, columns=df.columns)
# df0
```
